### Activity 1: Select only columns year, municipality ID, note for Math, note for Portuguese and ideb (`ano`, `id_municipio`, `nota_saeb_matematica`, `nota_saeb_lingua_portuguesa` e `ideb`) from Ideb, only for the year of 2019 and grouped by the municipalities.

```python
# importing library and files
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

ideb = pd.read_csv('/content/drive/MyDrive/ideb.csv')
pib = pd.read_csv('/content/drive/MyDrive/pib.csv')
```

```python
# selecting columns
ideb_colunas_select = ideb[['ano', 'id_municipio', 'nota_saeb_matematica', 'nota_saeb_lingua_portuguesa', 'ideb']]
ideb_colunas_select
```

```python
# creating table with the selected columns and filtering by year 2019
ideb_tab = ideb_colunas_select.loc[ideb_colunas_select['ano'] == 2019]
ideb_tab
```

```python
# grouping by id_municipality
ideb_final = ideb_tab.groupby('id_municipio').mean()
ideb_final
```

### Activity 2: Select the columns year, municipality ID, GDP and net tax (`ano`, `id_municipio`, `pib` e `impostos_liquidos`) from PIB, for the year of 2019 and grouping by municipality.

```python
# selecting columns
pib_colunas_select = pib[['ano', 'id_municipio', 'pib', 'impostos_liquidos']]
ideb_colunas_select
```

```python
# creating table with the selected columns and filtering by year 2019
pib_tab = pib_colunas_select.loc[pib_colunas_select['ano'] == 2019]
pib_tab
```

```python
# grouping by municipality ID
pib_final = pib_tab.groupby('id_municipio').mean()
pib_final
```

### Activity 3: Merge both DataFrames in one by the variable municipality ID.

```python
dados = pd.concat([ideb_final, pib_final], axis=1, join='outer')
dados
```

### Activity 4:  Check if there are missing data and check if there are irrelevant columns. If so, drop it.

```python
# showing DataFrame's information
dados.info()
```

**Answer**: There are 21 missing data and two irrelevant columns (ano and ano), as all data come from the year 2019.

```python
# droping irrelevant information
dados.drop(['ano'], inplace=True, axis=1)
dados
```

### Activity 6: Observing the variables in question, do you think there might be a correlation between municipalities with high Ideb score and a high GDP?

```python
# performing correlation analysis with Pearson
dados[['ideb', 'pib']].corr(method='pearson')
```

**Comment**: The result of the correlation analysis between ideb and pib is 0.03105. According to the Pearson correlation coefficient (r) table, when the result is near 0, it means that there is no correlation between the analyzed variables.

Source: https://www.statisticshowto.com/probability-and-statistics/correlation-coefficient-formula/

**Answer**: The result is near 0, thus there is no correlation between the mentioned variables.

```python
# using scatterplot to confirm that the correlation does not exist

import seaborn as sns

sns.scatterplot(data=dados, x='ideb', y='pib');
```

**Comment**: Scatterplots show the relationship between two variables. When the points follow a direction in the graph, then a relationship between the variable exists. If not, no relationship exists.

Source: https://www.khanacademy.org/math/statistics-probability/describing-relationships-quantitative-data/introduction-to-scatterplots/a/scatterplots-and-correlation-review

### What does the correlation between the variables `id_municipio` and `ideb` mean?

### Activity 8: Select the data only for the year of 2020 and only from the columns month, homicide 1, bodily injury with death, armed robbery, homicide by police intervention and drug trafficking (`mês`, `hom_doloso, lesao_corp_morte, latrocinio, cvli, hom_por_interv_policial`, e `trafico_drogas`).

```python
# selecting the columns
dados_rj = seg_rj[['mes', 'hom_doloso', 'lesao_corp_morte', 'latrocinio', 'cvli', 'hom_por_interv_policial', 'trafico_drogas']].loc[seg_rj['ano'] == 2020]
dados_rj
```

```python
# using scatterplot to see if there is a linear correlation between armed robbery and drug trafficking
sns.scatterplot(data=dados_rj, x='trafico_drogas', y='latrocinio');
```

There is no linear relationship between variables, thus Pearson should not be used.

```python
# using spearman to determine correlation
dados_rj[['latrocinio', 'trafico_drogas']].corr(method='spearman')
```

Correlation between armed robbery and drug trafficking is weak.



```python
sns.scatterplot(data=dados_rj, x='trafico_drogas', y='hom_doloso');
```

```python
dados_rj[['hom_doloso', 'trafico_drogas']].corr(method='spearman')
```

There is a moderate negative correlation between homicide 1 and drug trafficking: the higher the drug trafficking rate is, the lower homicide 1 rate is.

```python
sns.scatterplot(data=dados_rj, x='trafico_drogas', y='hom_por_interv_policial');
```

```python
dados_rj[['hom_por_interv_policial', 'trafico_drogas']].corr(method='spearman')
```

The correlation between the drug trafficking rate and the homicide by police intervention rate is weak.

```python
# using heatmap to see the correlation between all the variables
sns.heatmap(dados_rj.corr(), vmin=-1, vmax=1, annot=False);
```

**Comment**: The heatmap only uses variables that are numbers. Varible month, despite of being a number is not a continuous variable and makes no difference in the interpretation of the data. For this reason, the type of this variable was changed from int64 to object, so the map does not show it.

```python
# checking the types of variables
dados_rj.info()
```

```python
# changing the type of variable month (`mês`)
dados_rj = dados_rj.astype({'mes':'object'})
```

```python
# redoing the heatmap without the variable month
sns.heatmap(dados_rj.corr(), vmin=-1, vmax=1, annot=False);
```

**Answer**:
In general, there is a very strong positive correlation between the homicide 1 rate and the rate of intentional lethal and violent crimes (cvli), but this is just a matter of concept, since intentional homicide is an intentional lethal and violent crime.

There is also a positive correlation between cvli and the armed robbery rate, but it is weaker. This is possibly because the intentionality factor is less strong than in the previous case.

There is also a moderate positive correlation between cvli and hom_per_interv_policial rate.

There is a moderate negative correlation between homicide 1 and drug trafficking: the higher the drug trafficking rate, the lower the homicide 1 rate. This correlation also occurs with regard to the cvli rate.

There is no correlation (or it is a very weak one) between the rest of the variables.
